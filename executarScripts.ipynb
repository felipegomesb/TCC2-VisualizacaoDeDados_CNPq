```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import subprocess
import sys
```

```python
SCRIPTS = [
    # hierarquicas
    "hierarquicas/sunburst.py",
    "hierarquicas/treemap.py",

    # coropleticos
    "mapasCoropleticos/mapa_coropletico_nacional_somatoria.py",
    "mapasCoropleticos/mapa_coropletico_somatoria_internacional.py",
    "mapasCoropleticos/mapa_coropletico.py",
    "mapasCoropleticos/mapa_nacional_grandeArea.py",
    "mapasCoropleticos/mapa_coropletico_internacional_grandearea.py",
    "mapasCoropleticos/mapa_coropletico_internacional.py",

    # sankeyplot
    "sankeyPlot/sankeyplot_anoAGRUPADO.py",
    "sankeyPlot/sankeyplot_anoBOTAO.py",
    "sankeyPlot/sankeyplot_anoMODALIDADE.py",
    "sankeyPlot/sankeyplot_testes.py",

    # streamgraph
    "streamGraphs/streamgraph_plotly.py",

    # textual
    "visualizacaoTextual/wordcloud_palavraChave.py",
    "visualizacaoTextual/wordcloud_Titulo.py",
    "visualizacaoTextual/wordcloud_unificado.py",
]
```

```python
def executar_script(script):
    
    resultado = subprocess.run(
        [sys.executable, script],
        capture_output=True,
        text=True
    )

    if resultado.returncode == 0:
        print(f"[OK] {script}")
    else:
        print(f"[ERRO] {script}")
        print(resultado.stderr)

if __name__ == "__main__":

    # ajuste max_workers conforme seu PC
    with ThreadPoolExecutor(max_workers=6) as executor:

        futures = [
            executor.submit(executar_script, script)
            for script in SCRIPTS
        ]

        for future in as_completed(futures):
            future.result()
```
